```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

**Mengimport berbagai library yang di butuhkan untuk memvisualisasi data**

```python
df = pd.read_csv("GlobalWeatherRepository.csv")
```

```python
df.head(10)
```

_Data diatas hanya untuk menampilkan 10 data teratas pada dataset yang ada_

```python
cuaca = df["condition_text"].value_counts()
```

_Cuaca sebagai variabel yang menyimpan data frame berdasarkan kolom condition_text lalu menghitung berapa banyak target yang didapatkan_

__Menampilkan banyaknya data dari suatu target__

```python
cuaca
```

## Diagram batang
Menampilkan sebarapa banyak dan membandingkan curah cuaca yang didapatkan dari data set yang ada

```python
mean_data = df.groupby("condition_text")[["temperature_celsius","air_quality_Ozone"]].mean()
mean_data
```

```python
x = mean_data.index
y = mean_data["air_quality_Ozone"]
plt.bar(x, y)
plt.xticks(rotation = 90)
plt.title('Rata-rata Kualitas udara di lapisan ozone berdasarkan cuaca')
plt.xlabel("Cuaca")
plt.ylabel("Mean kualitas udara")
plt.show()
```

_Kesimpulan yang didapatkan dari visualisasi data diatas adalah tinkgkat kualitas udara di lapisan ozon pada setiap cuaca memiliki rentang yang berbeda dan didapatkan bahwa cuaca sangat cerah (sunny) memiliki kualitas udara yang paling bagus_

## PIE CHART
**Kebanyakan data bang**<br>
_dengan menggunakan explode PIE memecah pie chart yang ada agar lebih mudah di baca dalam bentuk visualisasi data_

```python
explode_pie = [0.1, 0.1, 0.7, 0.8, 0.9, 1, 1.1, 1.3, 1.5, 1.7, 1.9, 2.1, 2.3, 2.5, 2.7, 2.9, 3.1, 3.3, 3.5, 3.7, 3.9, 4.1]
plt.pie(cuaca, labels=cuaca.index,explode=explode_pie, startangle=90, autopct = "%.5f%%")
plt.show()
```

## HEATMAP pada visualisasi kualitas unsur udara
cor_matrix berfungsi untuk menampung data dalam bentuk matrik lalu di visualisasik menggankan sns yang merupakan seaborn agar dapat di tampilkan berdasarkan petak suhu agar di dapat kan penyebaran dan korelasi hubungan antar data

```python
cor_matrix = df.iloc[:, 29:35].corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```

_Pada visualisasi heatmap diatas didapat kan unsur yang terdapat pada udara mempengaruhi kualitas udara yang ada_

## LINE PLOT
Dengan menggunakan variabel penampung dan menggunakan fungsi .dt.date untuk mengambil tanggal nya saja lalu, dengan variabel kualitas agar didapatkan rata rata dari sekian banyak data pada tanggal yang sama, sehingga didapatkan data yang bisa di rubah menjadi sebuah line plot

```python
waktu = pd.to_datetime(df.last_updated).dt.date
kualitas = df.groupby(waktu)[["air_quality_Ozone"]].mean()
kualitas
```

```python
plt.figure(figsize=(10, 6))
x = kualitas.index
y = kualitas["air_quality_Ozone"]
plt.plot(x, y)
plt.xticks(rotation=60, horizontalalignment='right', fontsize=10)
plt.show()
```

_Pada rentang tanggal 29 agustus hingga sebelum tanggal 1 bulan september didapatkan bahwa terjadi penurunan kualitas udara yang cukup signifikan, lalu pada tanggal 13 bulan september terjadi kenaikan lalu mengalami penurunan kembali kualitas udara pada lapisan ozon sebelum tanggal 17 September 2023_
