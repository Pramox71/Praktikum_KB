## Pada Kali Ini akan menampilkan  jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
Lalu di masukkan kedalam dictionary agar mudah di tampilkan dalam bentuk record yang rapi

```python
record = df.shape[0]
atribut = df.columns
objek = df.select_dtypes(include=object).columns
angka = df.select_dtypes(include=np.number).columns
```

### Fungsi di atas

df.shape[] : fungsi untuk menampilkan ukuran baris(record) atau kolom(atribut) dari suatu dataframe

df.columns : fungsi untuk menampilkan nama-nama atribut atau kolom pada dataframe

df.select_dtypes() : fungsi untuk mengembalikan dataframe baru yang menyertakan kolom dari tipe yang ditentukan

include = object : untuk memilih semua tipe object 

include = np.number : untuk memilih semua tipe numerik/angka

df.isna().any() : untuk memeriksa apakah ada atribut yang memiliki nilai null

```python
info = {
    'jumlah_record' : record,
    'nama_atribut' : atribut,
    'atribut_object' : objek,
    'atribut_angka' : angka,
}

print("""
================================================
        A N A L I S I S   D A T A S E T
================================================""")
print("Jumlah Record \t\t\t\t:", info['jumlah_record'])

print("================================================")
print("Jumlah Atribut \t\t\t\t:", len(info['nama_atribut']))
print("================================================")
print("------------------------")
print("  No\t| Nama Atribut")
print("------------------------")
for i in range (len(info['nama_atribut'])):
    print(" ", i, "\t|", info['nama_atribut'][i])
    i+=1
    
print("\n================================================")
print("Jumlah Atribut dengan Tipe Objek \t:", len(info['atribut_object']))
print("================================================")
print("-------------------------------")
print("  No\t| Atribut Tipe Object")
print("-------------------------------")
for i in range (len(info['atribut_object'])):
    print(" ", i, "\t|", info['atribut_object'][i])
    i+=1
    
print("\n================================================")
print("Jumlah Atribut dengan Tipe Angka \t:", len(info['atribut_angka']))
print("================================================")
print("------------------------------")
print("  No\t| Atribut Tipe Angka")
print("------------------------------")
for i in range (len(info['atribut_angka'])):
    print(" ", i, "\t|", info['atribut_angka'][i])
    i+=1
print("\n================================================")
```

## Menampilkan Total nilai yang bertipe Object

```python
atribut_objek = df.select_dtypes(include='object')
```

_atribut_objek_ sebagai variabel penampung untuk menyeleksi data yang bertipe object

```python
for kolom in atribut_objek.columns:
    nilai_jumlah = atribut_objek[kolom].value_counts()
    print(f"Jumlah nilai berdasarkan label dari {kolom}:")
    print(nilai_jumlah)
    print("\n")
```

Lalu pada bagian diatas di kumpulkan sebuah data dan di totalkan pada kolom apa, dengan record yang sama, yaa intinya dalam satu kolom itu punya record apa aja dan total record yang sama ada berapa
